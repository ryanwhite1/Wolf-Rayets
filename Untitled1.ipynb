```python
# n = 256
@jit
def smooth_histogram2d(particles, weights, stardata):
    im_size = pixel_num
    
    x = particles[0, :]
    y = particles[1, :]
    
    xbound, ybound = jnp.max(jnp.abs(x)), jnp.max(jnp.abs(y))
    bound = jnp.max(jnp.array([xbound, ybound])) * (1. + 2. / im_size)
    
    xedges, yedges = jnp.linspace(-bound, bound, im_size+1), jnp.linspace(-bound, bound, im_size+1)
    return gm.smooth_histogram2d_base(particles, weights, stardata, xedges, yedges, im_size)
@jit
def smooth_histogram2d_w_bins(particles, weights, stardata, xbins, ybins):
    im_size = pixel_num
    return gm.smooth_histogram2d_base(particles, weights, stardata, xbins, ybins, im_size)
```

```python
n_params = 30
sq_err = np.zeros((n_params, n_params))
phases = np.linspace(0, 1, n_params)
periods = np.linspace(120, 240, n_params)
for j, period in enumerate(periods):
    apep_epoch = wrb.apep.copy()
    apep_epoch['period'] = period
    for i, phase in enumerate(phases):
        for year in vlt_years[1:]:
            data = vlt_data[year][0].data
            data /= np.max(data)

            apep_epoch['phase'] = phase + (year - 2016)/apep_epoch['period']
            particles, weights = model_func(apep_epoch)
            X, Y, H = smooth_histogram2d_w_bins(particles, weights, apep_epoch, pixels, pixels)

            H = np.array(H)
            sq_err[j, i] += np.sum((data - H)**2)
```

```python
fig, ax = plt.subplots()
X1, X2 = np.meshgrid(periods, phases)
pcm = ax.pcolormesh(X1, X2, sq_err.T, norm='log')
fig.colorbar(pcm)
```
