Now define our blackbody curve functions so that we can fit a curve to it.

```python
h = 6.626e-34
kb = 1.38e-23
b = 2.89777e-3
c = 299792458

jwst_lambdas = np.array([7.7e-6, 15e-6, 25.5e-6])

def weins_temp(wavelength):
    return b / wavelength
def blackbody(x, T):
    return (2 * h * c**2 / (x**5)) * 1 / (np.exp(h * c / (x * kb * T)) - 1)
def blackbody_mult(x, a, T):
    '''Blackbody curve including a scalar multiplier so that we can effectively non-dimensionalise/normalise it.'''
    return a * blackbody(x, T)

jwst_temps = weins_temp(jwst_lambdas)
print(jwst_temps)
```

```python
from scipy.optimize import curve_fit


# normalise the brightness of our pixels to a representative background pixel value in the image
shell_2_brightness = np.array([brightness_770_shell_2 / baseline770, brightness_1500_shell_2 / baseline1500, brightness_2550_shell_2 / baseline2550])
# shell_2_brightness = np.array([brightness_1500_shell_2 / baseline1500, brightness_2550_shell_2 / baseline2550])
shell_3_brightness = np.array([brightness_1500_shell_3 / baseline1500, brightness_2550_shell_3 / baseline2550])

fig, ax = plt.subplots()
ax.scatter(jwst_lambdas[1], shell_3_brightness[0], c='r', label='Shell 3 brightness vs wavelength')
ax.scatter(jwst_lambdas[2], shell_3_brightness[1], c='r')
ax.scatter(jwst_lambdas, shell_2_brightness, c='g', label='Shell 2 brightness vs wavelength')
# ax.scatter(jwst_lambdas[1], shell_2_brightness[0], c='g')
# ax.scatter(jwst_lambdas[2], shell_2_brightness[1], c='g', label='Shell 2 brightness vs wavelength')
ax.legend()
```

```python
fit_filters = np.array([jwst_lambdas[1], jwst_lambdas[2]])
# [shell_2_a, shell_2_T], _ = curve_fit(blackbody_mult, fit_filters, shell_2_brightness, p0=(1e-7, 50))
[shell_2_a, shell_2_T], _ = curve_fit(blackbody_mult, jwst_lambdas, shell_2_brightness, p0=(1e-7, 250))
[shell_3_a, shell_3_T], _ = curve_fit(blackbody_mult, fit_filters, shell_3_brightness, p0=(1e-8, 250))

fig, ax = plt.subplots()
ax.scatter(jwst_lambdas[1], shell_3_brightness[0], c='r')
ax.scatter(jwst_lambdas[2], shell_3_brightness[1], c='r')
# ax.scatter(jwst_lambdas[1], shell_2_brightness[0], c='g')
# ax.scatter(jwst_lambdas[2], shell_2_brightness[1], c='g')
ax.scatter(jwst_lambdas, shell_2_brightness, c='g')

# lambdas = np.linspace(1e-5, 3e-5, 100)
lambdas = np.linspace(0.7e-5, 3e-5, 100)

temps_shell_2 = blackbody_mult(lambdas, shell_2_a, shell_2_T)
temps_shell_3 = blackbody_mult(lambdas, shell_3_a, shell_3_T)
ax.plot(lambdas, temps_shell_2, color='tab:green', label=f'Shell 2: {shell_2_T:.2f}K')
ax.plot(lambdas, temps_shell_3, color='tab:red', label=f'Shell 3: {shell_3_T:.2f}K')
ax.set(xlabel='Wavelength (m)', ylabel='Estimated Brightness', title='Blackbody Curve Fits to Raw Images')
ax.legend()
```
